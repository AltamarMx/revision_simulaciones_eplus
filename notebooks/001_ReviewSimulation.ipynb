````python
import pandas as pd
import matplotlib.pyplot as plt
from iertools.read import read_sql
from dateutil.parser import parse
````

## Carga archivos del espacio de trabajo

````python
f = '../osm/CE_CASA 3D_270224/run/eplusout.sql'
tmp = read_sql(f,alias=True)
data = tmp.data
columnas = data.columns
columnas
````

````python
#| label: fig-scs
tmp.get_construction(tmp.construction_systems)
````

En la variable ```columnas``` se encuentran los nombres de las variables disponibles en los datos.

Con la siguiente celda, filtro de la variable ```columnas``` aquellas que contienen ```Ti_```, que son las temperaturas del aire al interior de cada zona t'ermica, y las guardo en la variable ```Tis```.

## Visualizaci'on de las temperaturas de las zonas t'ermicas 
Para poder visualizar esto tengo que pedirle a OpenStudio que me reporte la ```Zone Mean Air Temperature```. 

````python
Tis = [columna for columna in columnas if "Ti_" in columna]
Tis
````

Aqu'i voy a graficar todas las variables contenidas en ```Tis``` y en ```f1``` y ```f2``` especifico el rango de fechas. Por default la simulaci'on contiene todo un a~no de datos. Observa como defino la fecha final de la visualizaci'on ```f2``` en funci'on de la fecha inicial y m'as un periodo de tiempo de siete d'ias, en este caso. 

Para la primera semana de enero no se aprecia que exista alg'un termostato en funcionamiento, como se puede ver en la @fig-tis.

````python
#| label: fig-tis
#| fig-cap: Temperatura de cada una de las zonas t'ermicas.
Tis = [columna for columna in columnas if "Ti_" in columna]

fig, ax = plt.subplots(figsize=(12,4))
f1 = parse('2006-01-01')
f2 = f1 + pd.Timedelta('7D')
for Ti in Tis:
    ax.plot(data[Ti],label=Ti)

ax.set_xlim(f1,f2)
ax.legend();
````
